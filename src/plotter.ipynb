```python
from pathlib import Path
import numpy as np 
import pandas as pd
import seaborn as sns
sns.set()
```

```python
log_csv_file_path = Path.joinpath(Path.cwd(), 'data', 'log.csv')
output_fig_dir    = Path.joinpath(Path.cwd(), '..', 'final', 'img', 'result_img')
output_fig_dir.mkdir(parents=True, exist_ok=True)
```

```python
df = pd.read_csv(log_csv_file_path)
df.columns
```

```python
sns_plot = sns.lineplot(data=df, x='Episode', y='Mean Reward')
sns_plot.figure.savefig(Path.joinpath(output_fig_dir, 'mean_reward.png'))
```
